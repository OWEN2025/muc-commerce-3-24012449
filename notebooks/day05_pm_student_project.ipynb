### 专题分析记录

**数据现象：**  
不同生命周期阶段（TenureGroup）的用户，用户规模、平均订单量、返现金额、流失率存在明显差异：新用户群体平均返现金额偏低、流失率显著高于老用户群体；成熟老用户的下单频次更稳定、返现福利领取更多，用户留存表现更好。

**可能解释：**  
1.新用户对平台使用熟悉度不足，下单决策更谨慎，可能导致订单量与返现参与度偏低，流失风险更高；
2.老用户积累了平台使用经验与消费习惯，和高频下单、高返现领取行为呈相关关系，留存意愿更强；
3.不同生命周期阶段对应的运营扶持力度有区别，该因素对用户行为的影响需验证，可后续结合平台运营活动数据进一步分析确认。

## 任务4：双维度交叉分析（必做）

```python
# TODO：从以下维度中选择两个
# TenureGroup、Complain、PreferedOrderCat、CityTier、PreferredLoginDevice
dim_1 = None
dim_2 = None

# TODO：按两个维度统计用户数、流失人数、流失率，以及至少一个行为指标
cross_analysis = None

# TODO：新增“样本提示”列；用户数<30标记为“小样本”，否则为“可观察”

# TODO：按流失率或用户数排序并展示
# display(cross_analysis)

# TODO 1：选定两个分析维度
dim_1 = "TenureGroup"
dim_2 = "CityTier"

# TODO 2：双维度分组聚合，指标：用户数、流失人数、流失率、平均订单数（行为指标）
cross_analysis = df.groupby([dim_1, dim_2]).agg(
    用户数=("CustomerID", "nunique"),
    流失人数=("Churn", "sum"),
    平均订单数=("OrderCount", "mean")
).reset_index()

# 计算流失率
cross_analysis["流失率"] = cross_analysis["流失人数"] / cross_analysis["用户数"]

# TODO 3：新增样本提示列，用户数<30标记小样本
cross_analysis["样本提示"] = cross_analysis["用户数"].apply(lambda x: "小样本" if x < 30 else "可观察")

# TODO 4：按流失率降序排序展示
cross_analysis = cross_analysis.sort_values(by="流失率", ascending=False)

# 展示结果
display(cross_analysis)
```

```python
# 检查点4
assert dim_1 in df.columns and dim_2 in df.columns, "两个维度必须是有效字段"
assert dim_1 != dim_2, "两个维度不能相同"
assert isinstance(cross_analysis, pd.DataFrame), "cross_analysis应为DataFrame"
assert {"用户数", "流失率", "样本提示"}.issubset(cross_analysis.columns), "双维表缺少必需列"
assert set(cross_analysis["样本提示"]).issubset({"小样本", "可观察"}), "样本提示取值不正确"
print("检查点4通过")
```

## 任务5：报表输出与回读验证（必做）

```python
# TODO：将三个表导出到OUTPUT_DIR
# 文件名必须为：overall_metrics.csv、segment_analysis.csv、cross_analysis.csv
# 要求：index=False，encoding="utf-8-sig"

outputs = {
    "overall_metrics.csv": overall_metrics,
    "segment_analysis.csv": segment_analysis,
    "cross_analysis.csv": cross_analysis,
}

# TODO：循环导出并重新读取；打印每个文件的shape

# 定义需要导出的三张表映射
outputs = {
    "overall_metrics.csv": overall_metrics,
    "segment_analysis.csv": segment_analysis,
    "cross_analysis.csv": cross_analysis,
}

# 循环导出CSV，再回读验证并打印shape
for filename, df_table in outputs.items():
    # 拼接完整输出路径
    save_path = OUTPUT_DIR / filename
    # 导出：index=False，编码utf-8-sig
    df_table.to_csv(
        save_path,
        index=False,
        encoding="utf-8-sig"
    )
    print(f"✅ 已导出文件：{save_path.as_posix()}")
    
    # 回读文件
    read_df = pd.read_csv(save_path, encoding="utf-8-sig")
    # 打印文件shape
    print(f"📄 回读 {filename} 数据行列：{read_df.shape}")
    print("-" * 50)
```

### 结论2

在全量分层用户里，生命周期越长的群体，平均订单数越高、流失率越低；7-12 个月、13-24 个月、24 个月以上成熟用户平均订单均稳定在 3 单以上，留存表现显著优于新用户。对应证据表：segment_analysis（单维度生命周期分层表）。

二线城市新用户样本仅 15 人（小样本），流失率高达 93%，三线城市新用户样本充足共 181 人，流失率 57%；同生命周期下城市层级越高，新用户流失风险越低。对应证据表：cross_analysis（双维度交叉分析表）。

### 分析限制

1.数据集缺少订单金额、下单时间字段，无法计算 GMV、客单价、复购周期时序趋势，只能基于订单计数做行为判断；

2.缺少用户渠道来源、平台营销活动记录，无法区分高流失新用户是产品适配问题还是获客质量问题；

3.部分细分人群（二线城市新用户、24 个月以上二线城市用户）样本量小于 30 属于小样本，数据波动大，结论代表性有限。

### 运营建议与验证方式

运营时，针对三线城市新用户群体推出新人专属优惠券、简化下单操作引导，降低该人群前期流失风险。

1.拆分人群做 A/B 对照实验：一组发放新人券，一组不发放；

2.补充追踪 30 天内两组人群的留存率、订单数量、流失率；

3.需要新增营销活动记录表、用户下单时间序列数据，对比干预前后用户行为变化，判断策略是否有效降低流失。
